## Selenium

```python
import requests
from bs4 import BeautifulSoup
import os
import time 

for page in range(18, 41):

    # URL of the page to scrape
    url = f'https://www.plasticsurgery.org/photo-gallery/procedure/facelift/page/{page}'
    print(url)
    # Send a GET request to the URL
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the main div containing the images
    gallery_div = soup.find('div', class_='gallery grid-container Component')

    # Recursively find all img tags within the main div
    image_urls = [img['src'] for img in gallery_div.find_all('img', class_='procedure')]

    # Print the extracted image URLs
    print(image_urls)

    # Directory to save the downloaded images
    save_dir = 'downloaded_images'
    os.makedirs(save_dir, exist_ok=True)

    # Fåunction to download an image
    def download_image(url, save_path):
        img_response = requests.get(url)
        if img_response.status_code == 200:
            with open(save_path, 'wb') as file:
                file.write(img_response.content)

    # Download each image
    for i, img_url in enumerate(image_urls):
        img_name = f'image_{page}_{i+1}.jpg'
        save_path = os.path.join(save_dir, img_name)
        download_image(img_url, save_path)
        print(f'Downloaded {img_name}')

    time.sleep(1)
```
