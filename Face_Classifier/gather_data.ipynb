```python
import shutil
import os

# Paths to the big folders
source_folders = ["./Dataset_Solo", "./Dataset"]
destination_folder = "./Dataset_Mixed"

# Subfolders to copy
subfolders = ["Angle", "Front", "Side"]

# Function to copy files from source to destination
def copy_files(src, dst):
    for item in os.listdir(src):
        s = os.path.join(src, item)
        d = os.path.join(dst, item)
        if os.path.isfile(s):
            shutil.copy2(s, d)
            print(f'Copied {s} to {d}')

# Iterate over each subfolder
for subfolder in subfolders:
    for source_folder in source_folders:
        src_path = os.path.join(source_folder, subfolder)
        dst_path = os.path.join(destination_folder, subfolder)
        copy_files(src_path, dst_path)

print("Files copied successfully.")
```

```python
import requests
from bs4 import BeautifulSoup

case_links = []

for page in range(1, 41):
    # URL of the page to scrape
    url = f'https://www.plasticsurgery.org/photo-gallery/procedure/facelift/page/{page}'
    
    # Send a GET request to the URL
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all <a> tags with the class 'subgrid' and extract the href attribute
    links = [a['href'] for a in soup.find_all('a', class_='subgrid')]

    # Add the extracted links to case_links
    case_links.extend(links)

# Print all collected links
print(case_links)
```

```python
import requests
from bs4 import BeautifulSoup
import os
import time 

for idx, url in enumerate(case_links):

    # URL of the page to scrape
    # url = f'https://www.plasticsurgery.org/photo-gallery/procedure/facelift/page/{page}'
    # print(url)
    # Send a GET request to the URL
    print(url)
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    image_tags = soup.find_all('a', class_='gallery-case')
    image_urls = [img['href'] for img in image_tags]
    
    # Print the extracted image URLs
    print(image_urls)

    # Directory to save the downloaded images
    save_dir = 'downloaded_images'
    os.makedirs(save_dir, exist_ok=True)

    # Fåunction to download an image
    def download_image(url, save_path):
        img_response = requests.get(url)
        if img_response.status_code == 200:
            with open(save_path, 'wb') as file:
                file.write(img_response.content)

    # Download each image
    for i, img_url in enumerate(image_urls):
        img_name = f'image_{idx}_{i+1}.jpg'
        save_path = os.path.join(save_dir, img_name)
        download_image(img_url, save_path)
        print(f'Downloaded {img_name}')

    time.sleep(1)
```
